```python
import random
import os
from pathlib import Path
```

```python
v1a_path = 'data4v1a/train/images'
v1a_label_path = 'data4v1a/train/labels'
nv_files = [f.name for f in Path(v1a_path).iterdir() if f.name.startswith('NV_')]
v_files = [f.name for f in Path(v1a_path).iterdir() if f.name.startswith('V_')]
```

```python
nv_files.sort()
len(nv_files), nv_files[:10]
```

```python
smart_school_path = 'smart-schools-ai/train/images'
smart_school_label_path = 'smart-schools-ai/train/labels'

for nv_file in nv_files:
    print(f'Processing {nv_file} ...')
    prefix = nv_file.split('.')[0]
    ss_files = [f.name for f in Path(smart_school_path).glob(f"{prefix}*")]
    ss_files = [f for f in ss_files if f != nv_file][:2]
    for ss_nv_file in ss_files:
        os.system(f'cp {smart_school_path}/{ss_nv_file} {v1a_path}')
        os.system(f'cp {smart_school_label_path}/{ss_nv_file.rsplit(".", 1)[0]}.txt {v1a_label_path}')
```

```python
v_files_rm = random.sample(v_files, 900)
for v_file in v_files_rm:
    os.system(f'rm -f {v1a_path}/{v_file}')
    os.system(f'rm -f {v1a_label_path}/{v_file.rsplit(".", 1)[0]}.txt')
    print(v_file, 'removed.')
```

```python
v1a_test_path = 'data4v1a/test'
smart_school_path = 'smart-schools-ai/images'
smart_school_label_path = 'smart-schools-ai/labels'
nv_test_files = [f.name for f in Path(v1a_test_path+'/images').iterdir() if f.name.startswith('NV_')]

for nv_file in nv_test_files:
    prefix = nv_file.split('.')[0]
    ss_files = [f.name for f in Path(smart_school_path).glob(f"{prefix}*")]
    ss_files = [f for f in ss_files if f != nv_file][:3]
    print(f'Processing {nv_file} ({len(ss_files)} replicas) ...')
    
    for ss_nv_file in ss_files:
        os.system(f'cp {smart_school_path}/{ss_nv_file} {v1a_test_path}/images')
        os.system(f'cp {smart_school_label_path}/{ss_nv_file.rsplit(".", 1)[0]}.txt {v1a_test_path}/labels')
    # break
```

```python
v1a_val_path = 'data4v1a/valid'
smart_school_path = 'smart-schools-ai/images'
smart_school_label_path = 'smart-schools-ai/labels'
nv_val_files = [f.name for f in Path(v1a_val_path+'/images').iterdir() if f.name.startswith('NV_')]

for nv_file in nv_val_files:
    prefix = nv_file.split('.')[0]
    ss_files = [f.name for f in Path(smart_school_path).glob(f"{prefix}*")]
    ss_files = [f for f in ss_files if f != nv_file][-4:]
    print(f'Processing {nv_file} ({len(ss_files)} replicas) ...')
    
    for ss_nv_file in ss_files:
        os.system(f'cp {smart_school_path}/{ss_nv_file} {v1a_val_path}/images')
        os.system(f'cp {smart_school_label_path}/{ss_nv_file.rsplit(".", 1)[0]}.txt {v1a_val_path}/labels')
    # break
```

```python
v_files = [f.name for f in Path(v1a_test_path+'/images').iterdir() if f.name.startswith('V_')]
v_files_rm = random.sample(v_files, 120)
for v_file in v_files_rm:
    os.system(f'rm -f {v1a_test_path}/images/{v_file}')
    os.system(f'rm -f {v1a_test_path}/labels/{v_file.rsplit(".", 1)[0]}.txt')
    print(v_file, 'removed.')
```
