```python
import random
import os
import json
import cv2
```

```python
ds_split = 'train'  # train or val

src_paths = [
    (f"godseye-violence-detection-dataset/{ds_split}/NonFight/", "no"),
    (f"godseye-violence-detection-dataset/{ds_split}/Fight/", "fi"),
]

dest_paths = (f'hockey-data/images/{ds_split}', f'hockey-data/labels/{ds_split}')
```

```python
files = []
for p in src_paths:
    files += [(p[0], f) for f in os.listdir(p[0]) if f.startswith(p[1])]

print(files[0])
```

```python
for file in files:
    src_path = f'{file[0]}/{file[1]}/{file[1]}'
    jpg_path = f"{dest_paths[0]}/{file[1]}.jpg"
    label_path = f"{dest_paths[1]}/{file[1]}.txt"
    
    with open(src_path + '.json') as f:
        frame_info = json.load(f)
    
    for i in random.sample(range(len(frame_info)), 10):
        objs = frame_info[i]['detections']
        if len(objs) > 0:
            print('video:', src_path+'.avi', i)
            # print('jpg:', jpg_path)
            # print('label:', label_path)
            cap = cv2.VideoCapture(src_path+'.avi')
            w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
            h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
            ret, frame = cap.read()
            cv2.imwrite(jpg_path, frame)
            cap.release()

            # for obj in objs:
            #     box = obj['box']
            #     x1, y1, x2, y2 = int(box['x1']), int(box['y1']), int(box['x2']), int(box['y2'])
            #     xx = (x1 + x2) / w / 2
            #     yy = (y1 + y2) / h / 2
            #     ww = (x2 - x1) / w
            #     hh = (y2 - y1) / h
            #     print(f"{file[1].startswith('fi')} {xx} {yy} {ww} {hh}")

            with open(label_path, 'w') as f:
                for obj in objs:
                    box = obj['box']
                    x1, y1, x2, y2 = int(box['x1']), int(box['y1']), int(box['x2']), int(box['y2'])
                    xx = (x1 + x2) / w / 2
                    yy = (y1 + y2) / h / 2
                    ww = (x2 - x1) / w
                    hh = (y2 - y1) / h
                    f.write(f"{int(file[1].startswith('fi'))} {xx} {yy} {ww} {hh}\n")

            break
```

```python
video_path = "godseye-violence-detection-dataset/val/Fight/0Ow4cotKOuw_4/0Ow4cotKOuw_4"
frame_no = 15
cap = cv2.VideoCapture(video_path+'.avi')
cap.set(cv2.CAP_PROP_POS_FRAMES, 15)
ret, frame = cap.read()
cap.release()
```

```python
with open(f'{video_path}.json') as f:
    frame_info = json.load(f)
    for detect in frame_info[frame_no]['detections']:
        box = detect['box']
        p1 = (int(box['x1']), int(box['y1']))
        p2 = (int(box['x2']), int(box['y2']))
        cv2.rectangle(frame, p1, p2, (0, 255, 0), 2)

cv2.imwrite(f"frame_{frame_no}.jpg", frame)
```
