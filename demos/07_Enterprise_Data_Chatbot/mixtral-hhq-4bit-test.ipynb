```python
from hqq.engine.hf import HQQModelForCausalLM, AutoTokenizer
```

```python
model_path = '/home/cdsw/models/Mixtral-8x7B-Instruct-v0.1-hf-4bit_g64-HQQ'
tokenizer = AutoTokenizer.from_pretrained(model_path)
```

```python
model = HQQModelForCausalLM.from_quantized(
    model_path,
    # use_cache=True,
    device='auto'
)
```

```python
model
```

```python
prompt = "<s> [INST] How do I build a car? [/INST] "

inputs = tokenizer(prompt, return_tensors="pt", add_special_tokens=False)
outputs = model.generate(**(inputs.to('cuda')), max_new_tokens=1000)
print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```
