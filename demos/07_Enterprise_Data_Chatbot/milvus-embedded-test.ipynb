```python
from milvus import default_server
from pymilvus import MilvusClient, DataType
# from pymilvus import connections, utility
```

```python
default_server.start()
```

```python
uri = "http://localhost:%s" % default_server.listen_port
client = MilvusClient(uri=uri)
```

```python
schema = MilvusClient.create_schema(
    auto_id=False,
    enable_dynamic_field=True,
)
```

```python
schema.add_field(field_name="id", datatype=DataType.INT64, is_primary=True)
schema.add_field(field_name="vector", datatype=DataType.FLOAT_VECTOR, dim=768)
```

```python
dir(DataType)
```

```python
index_params = client.prepare_index_params()
# index_params.add_index(field_name="id")
index_params.add_index(
    field_name="vector", 
    index_type="AUTOINDEX",
    metric_type="COSINE"
)
```

```python
client.create_collection(
    collection_name="rag4fin_consine",
    schema=schema,
    index_params=index_params
)
```

```python
client.list_indexes(collection_name="rag4fin_consine")
```

```python
client.describe_index(collection_name="rag4fin_consine", index_name="vector")
```
